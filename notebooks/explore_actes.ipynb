## Chargement des données

Commençons par charger les données dans pandas.

```python
from pathlib import Path

import pandas as pd
import pandera as pa

from pandera import Check, Column, DataFrameSchema
```

```python
CSV_RECHER = Path("../data/raw/recherche-actes-20220308-1126.csv")
CSV_EXPORT = Path("../data/raw/export-actes-20220308-1126.csv")
```

```python
# le début des fichiers contient 12 lignes non-vides de métadonnées de recherche
# les intitulés de colonnes sont donc sur la ligne 13 (index 12)
dtype_actes = {
    "N° Acte": "string",
    "Nature": "category",
    "Matière": "category",
    "Objet": "string",
    "État de l'acte": "category",
    "Caractère prioritaire": "category",
    "Mode de transmission": "category",
    "Acte contrôlé": "category",
    "Acte annulé": "category",
    "Service attributaire": "category",
    "Date de transmission du courrier simple": "string",  # TODO date?
    "Date de transmission de la lettre d'observation": "string",  # TODO date?
    "Date de demande de transmission complémentaire": "string",  # TODO date?
    "Date de saisie de l'expert": "string",  # TODO date?
    "Date Réponse Expert": "string",  # TODO date?
    "Nature réponse Expert": "string",  # TODO category? absence de valeurs
    "Libellé émetteur (ACTES)": "category",  # numéro d'arrondissement puis commune
    "Nature émetteur": "category",
    "Tiers de transmission": "category",
    "Département": "category",
    "Arrondissement": "category",
    "Région": "category",
    "Strate": "category",
}

df_recher = pd.read_csv(CSV_RECHER, encoding="iso-8859-15", sep=';', header=12, dtype=dtype_actes)
df_export = pd.read_csv(CSV_EXPORT, encoding="iso-8859-15", sep=';', header=12, dtype=dtype_actes)
```

## Tentative de définition d'un schéma de données et validation du contenu du fichier

Nous utilisons pandera.

```python
DFSCHEMA_RECHER = Path("../scripts/dfschema_actes_recher.py")
if not DFSCHEMA_RECHER.exists():
    # à exécuter une fois seulement, puis on itère sur le script généré dans scripts/
    schema_recher = pa.infer_schema(df_recher.rename(columns={x: x.replace("'", "\\'") for x in df_recher.columns}))
    schema_recher.to_script(DFSCHEMA_RECHER)
```

```python
DFSCHEMA_EXPORT = Path("../scripts/dfschema_actes_export.py")
if not DFSCHEMA_EXPORT.exists():
    # à exécuter une fois seulement, puis on itère sur le script généré dans scripts/
    schema_export = pa.infer_schema(df_export.rename(columns={x: x.replace("'", "\\'") for x in df_export.columns}))
    schema_export.to_script(DFSCHEMA_EXPORT)
```

## Comparaison des deux listes


```python
df_comp = pd.merge(df_recher, df_export, how="outer", indicator=True)
df_comp
```

```python
# on vérifie que l'export est bien un sous-ensemble des résultats de la recherche
assert df_comp[df_comp["_merge"] == "right_only"].empty
```

```python
# on examine les résultats jugés non pertinents par la DDTM
df_comp[df_comp["_merge"] == "left_only"][["Objet"]]
```
